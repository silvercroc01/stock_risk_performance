```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load daily returns
returns = pd.read_csv('../data/sector_returns.csv', index_col=0, parse_dates=True)
returns.tail()
```

```python
plt.figure(figsize=(15, 10))
returns.hist(bins=50, figsize=(15,10))
plt.suptitle("Distribution of Daily Returns by Sector")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(returns.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation of Sector Returns")
plt.show()
```

```python
rolling_vol = returns.rolling(window=21).std() * (252**0.5)
rolling_vol.plot(figsize=(12,6))
plt.title("21-Day Rolling Annualized Volatility")
plt.ylabel("Volatility")
plt.show()
```
